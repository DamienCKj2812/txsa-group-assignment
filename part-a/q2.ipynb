# Q2. Word Stemming

1. **Importance of Stemming in Text Analytics**


Stemming is a key technique in text pre-processing for NLP tasks. It reduces words to their root form, which helps to simplify data and enhance model performance. The main benefits of stemming include:

- **Reduces Word Variability:** Different forms of a word (e.g., "running", "runs", "ran") are reduced to a common root ("run"), minimizing vocabulary size and dimensionality.
- **Improves Efficiency:** By reducing the number of unique words, stemming makes text processing and analysis more efficient.
- **Enhances Matching and Retrieval:** Stemming helps match related words in search and retrieval tasks, improving the accuracy of information retrieval systems.
- **Improves Generalization:** Models trained on stemmed data can generalize better, as similar words are treated as the same feature.
- **Reduces Noise:** Stemming helps clean and normalize text, making downstream NLP tasks like sentiment analysis, classification, and clustering more robust.

Overall, stemming is essential for cleaning and standardizing text data, leading to improved performance in various NLP and text mining applications.

2. **Demonstrate Word Stemming**

```python
import re
from nltk.stem import PorterStemmer, LancasterStemmer

# Load Data_1.txt
file_path = '../assets/dataset-a/Data_1.txt'
with open(file_path, 'r') as file:
    text = file.read().strip()

# Split text into words
words = text.split()
print("Original Words:")
print(words)

# 1. Regex Stemmer (removes common suffixes)
def regex_stemmer(word):
    return re.sub(r'(ing|ly|ed|es|s|er)$', '', word)

regex_stems = [regex_stemmer(word) for word in words]
print("\nRegex Stemmer Output:")
print(regex_stems)

# 2. Porter Stemmer
porter = PorterStemmer()
porter_stems = [porter.stem(word) for word in words]
print("\nPorter Stemmer Output:")
print(porter_stems)

# 3. Lancaster Stemmer
lancaster = LancasterStemmer()
lancaster_stems = [lancaster.stem(word) for word in words]
print("\nLancaster Stemmer Output:")
print(lancaster_stems)
```

3. **Differences among Regex, Porter and Lancaster Stemmer**

| Stemmer               | Characteristics                                                    | Observations from Output                                                                                                                       |
| --------------------- | ------------------------------------------------------------------ | ---------------------------------------------------------------------------------------------------------------------------------------------- |
| **Regex Stemmer**     | Simple rule-based stemmer that removes common suffixes             | Produces inconsistent stems (`Classification -> Classification`, `choosing -> choos`, `class -> clas`). Does not always reduce words effectively. |
| **Porter Stemmer**    | Moderate rule-based algorithm for English; balances aggressiveness | Produces more consistent stems (`Classification -> classif`, `choosing -> choos`, `class -> class`). Keeps some word meaning intact.              |
| **Lancaster Stemmer** | Aggressive, short-stemmer algorithm                                | Produces the shortest stems (`Classification -> class`, `choosing -> choos`, `given -> giv`, `basic -> bas`). Can over-stem, losing some nuance.   |

- Regex Stemmer is customizable but naive.
- Porter Stemmer is widely used in NLP and balances accuracy and stemming aggressiveness.",
- Lancaster Stemmer is very aggressive; it reduces words more drastically, which can sometimes remove important meaning.",

**Examples of Usage:**
- **Regex Stemmer:** Useful for quick preprocessing in small projects or when you need to apply simple, custom rules (e.g., cleaning up user-generated content or prototyping).",
- **Porter Stemmer:** Commonly used in search engines, information retrieval, and general NLP pipelines where a balance between accuracy and reduction is needed.",
- **Lancaster Stemmer:** Suitable for applications where reducing words to their shortest root is important, such as spam detection or when working with very noisy text.",

Conclusion:
Regex is good for quick rules, Porter is good for general NLP tasks, and Lancaster is suitable when aggressive reduction is needed.",
Choice depends on the task goal, e.g., text classification vs. detailed linguistic analysis.
